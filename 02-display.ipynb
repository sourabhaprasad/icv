```python
import cv2
import os
import matplotlib.pyplot as plt

# Folder path containing the images
folder_path = 'images'  # Change this to your folder path

# Supported image formats
image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff')

# Get list of image files in the folder
image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(image_extensions)]

if not image_files:
    print(f"No image files found in folder: {folder_path}")
else:
    plt.figure(figsize=(15, 5 * len(image_files)))

    for idx, image_name in enumerate(image_files):
        image_path = os.path.join(folder_path, image_name)
        image = cv2.imread(image_path)

        if image is not None:
            image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            plt.subplot(len(image_files), 1, idx + 1)
            plt.imshow(image_rgb)
            plt.title(image_name)
            plt.axis('off')
        else:
            print(f"Could not load image: {image_name}")

    plt.tight_layout()
    plt.show()
```
